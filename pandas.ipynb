```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
sns.set_theme(style="whitegrid")
OUT_DIR = "plots"
os.makedirs(OUT_DIR, exist_ok=True)
```

```python
orange=pd.read_excel(r"C:\Users\adith\Downloads\archive (2)\Orange Cap.xlsx")
```

```python
orange["Average (Avg)"] = pd.to_numeric(orange["Average (Avg)"], errors="coerce")
```

```python
orange.columns = orange.columns.str.strip()
```

```python
top_batters = orange.nsmallest(10, "Rank")
plt.figure(figsize=(10, 6))
sns.barplot(data=top_batters, x="Runs", y="Player Name", hue="Team", dodge=False)
plt.title("Top 10 Run-Scorers (IPL 2025)")
plt.tight_layout()
plt.savefig(f"{OUT_DIR}/top10_run_scorers.png", dpi=150)
```
